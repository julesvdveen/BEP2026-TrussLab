Step 1: Bridge Design & Topology
</h3>
<p style="margin-bottom: 0; text-align: justify; opacity: 0.9; font-size: 14px;">
Now that you have inspected the hardware kit, you will begin translating a physical structural design into a formal discrete system matrix. Below, we review the fundamental governing equilibrium layout before setting your initial geometric parameters.
</p>
</div>

<div style="background: rgba(0, 166, 214, 0.03); border-left: 4px solid #00a6d6; padding: 10px 14px; border-radius: 0 4px 4px 0; border: 1px solid rgba(0, 166, 214, 0.1); margin-bottom: 25px; font-size: 13.5px;">
<strong style="color: #00a6d6;">📘 Theory:</strong> In a discrete 2D truss framework, the structural equilibrium is governed by the linear system equation <span style="font-family: 'Times New Roman', Times, serif; font-size: 15.5px;"><b><i>f</i> = <i>Kd</i></b></span>. 
<ul style="margin: 5px 0 0 0; padding-left: 20px; list-style-type: circle;">
    <li style="margin-bottom: 4px;"><strong>Global Stiffness Matrix (<span style="font-family: 'Times New Roman', Times, serif;"><i>K</i></span>):</strong> Assembled by mapping member stiffnesses and geometric transformation matrices. It defines the internal load paths of the entire assembly.</li>
    <li style="margin-bottom: 4px;"><strong>External Force Vector (<span style="font-family: 'Times New Roman', Times, serif;"><i>f</i></span>):</strong> Contains the external horizontal and vertical loads applied to each joint node.</li>
    <li style="margin-bottom: 0;"><strong>Joint Displacement Vector (<span style="font-family: 'Times New Roman', Times, serif;"><i>d</i></span>):</strong> Represents the directional translations of the nodes resulting from the applied loads.</li>
</ul>
</div>

📡 The measurements behind TrussLab
</h3>

<p style="margin-bottom: 12px; text-align: justify; opacity: 0.9; font-size: 14px;">
To measure how the bridge displaced, we look at the deformation of the individual beams. Each beam is given an electrical resistance that changes linearly with its length.
</p>

<p style="margin-bottom: 12px; text-align: justify; opacity: 0.9; font-size: 14px;">
By sending an electrical signal through the bridge, the electrical resistance can be determined from the voltage drop over a reference resistor. This makes it possible to estimate the length change of each beam.
</p>

<p style="margin-bottom: 12px; text-align: justify; opacity: 0.9; font-size: 14px;">
Because a bridge consists of multiple beams, the system contains several unknown variables. To separate these contributions, each beam is made part of an electrical filter, so that its resistance response also depends on frequency.
</p>

<p style="margin-bottom: 0; text-align: justify; opacity: 0.9; font-size: 14px;">
When a range of frequencies is swept through the bridge using the myDAQ, a Bode plot can be created. By analysing this Bode plot, the individual resistances can be obtained, and from these values the lengths of the beams can be determined.
</p>
